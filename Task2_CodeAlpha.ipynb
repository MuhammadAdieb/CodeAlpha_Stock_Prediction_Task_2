# Extract closing stock   of one company, here YHOO, Display Shape , Convert stock to 2D array

```python
display (stocks.shape)
```

```python
stocks = stocks.reshape(len(stocks) , 1)
print (stocks.shape)
print(stocks)
```

# Min Max Scaler

# Feature scaling the vector for better model performance.

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1)) 
stocks = scaler.fit_transform(stocks) 
display (stocks)
```

# Train Test Split (80 % Training and 20 % Test Data)     

```python
print (stocks.shape) 
train = int(len(stocks) * 0.80)
print (train)
test = len(stocks) - train 
print (test)
```

# Training Data

```python
train = stocks[0:train]
display (train.shape)
print(train)
```

# Test Data

```python
test = stocks[len(train) : ]
display(test.shape)
display (test)
```

# Function to prepare the data, use two previous values to predict for third data , Call the function to prepare training data, Call the function to prepare test data, Call the function to prepare full  data 

```python
def process_data(data , n_features):
    dataX, dataY = [], [] 
    for i in range(len(data)-n_features): 
        a = data[i:(i+n_features), 0] 
        dataX.append(a) 
        dataY.append(data[i + n_features, 0])
    return np.array(dataX), np.array(dataY)
```

```python
n_features = 2
trainX, trainY = process_data(train, n_features)
print(trainX.shape , trainY.shape)
```

```python
n_features = 2

testX, testY = process_data(test, n_features)
print (testX.shape , testY.shape)
```

```python
n_features = 2
stocksX, stocksY = process_data(stocks, n_features)
print (stocksX.shape , stocksY.shape)
```

```python
display (trainX[:10])
```

```python
display (trainY[:10])
```

# Reshape to 3D array to use for LSTM /Deep Learning model 

```python
trainX = trainX.reshape(trainX.shape[0] , 1 ,trainX.shape[1])
display (trainX.shape)
```

```python
testX = testX.reshape(testX.shape[0] , 1 ,testX.shape[1])
display (testX.shape)
```

```python
stocksX= stocksX.reshape(stocksX.shape[0] , 1 ,stocksX.shape[1])
display (stocksX.shape)
```

# Import Libraries

```python
import math  
from keras.models import Sequential 
from keras.layers import Dense , BatchNormalization , Dropout , Activation 
from keras.layers import LSTM , GRU 
from sklearn.preprocessing import MinMaxScaler 
from sklearn.metrics import mean_squared_error 
from keras.optimizers import Adam , SGD , RMSprop
```

# Creating checkpoints

```python
filepath="stock_weights1.hdf5"
from keras.callbacks import ReduceLROnPlateau , ModelCheckpoint
lr_reduce = ReduceLROnPlateau(monitor='val_loss', factor=0.1, epsilon=0.0001, patience=1, verbose=1)
checkpoint = ModelCheckpoint(filepath, monitor='val_loss', verbose=1, save_best_only=True, mode='max')
```

# Create the model â€“ Fully connected Layer 

```python
model = Sequential()
model.add(GRU(256 , input_shape = (1 , n_features) , return_sequences=True))
model.add(Dropout(0.4))
model.add(LSTM(256))
model.add(Dropout(0.4))
model.add(Dense(64 ,  activation = 'relu'))
model.add(Dense(1))
print(model.summary())
```

```python
model.compile(loss='mean_squared_error', optimizer=Adam(lr = 0.0005) , metrics = ['mean_squared_error'])
```

# Fit the model

```python
history = model.fit(trainX, trainY, epochs=100 , batch_size = 128 , 
          callbacks = [checkpoint , lr_reduce] , validation_data = (testX,testY))    
```

# Predict the value for testX, display top 10 records , Inverse Transform the data â€“ Convert the data to Original form, Reshape and Display Original Data 

```python
test_pred = model.predict(testX)
display (test_pred [:10])
```

```python
test_pred = scaler.inverse_transform(test_pred)
display (test_pred [:10])
```

```python
testY = testY.reshape(testY.shape[0] , 1)
testY = scaler.inverse_transform(testY)
display (testY[:10])
```

# Display the accuracy 

```python
from sklearn.metrics import r2_score
r2_score(testY,test_pred)
```

# Plot the Graph Actual and Predicted Data 

```python
print("Red - Predicted Stock Prices  ,  Blue - Actual Stock Prices")
plt.rcParams["figure.figsize"] = (15,7)
plt.plot(testY , 'b')
plt.plot(test_pred , 'r')
plt.xlabel('Time')
plt.ylabel('Stock Prices')
plt.title('Check the accuracy of the model with time')
plt.grid(True)
plt.show()
```
